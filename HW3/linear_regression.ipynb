```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math

from sklearn import datasets, linear_model, model_selection
from sklearn.metrics import mean_squared_error, r2_score
```

```python
dataset = datasets.fetch_california_housing()

print("Shape of Input: ", dataset.data.shape)
print("Shape of Target: ", dataset.target.shape)
print("First five rows of Input: \n", dataset.data[:5])
print("First five rows of Target: \n", dataset.target[:5])
```

```python
X_train, X_test, y_train, y_test = model_selection.train_test_split(dataset.data, dataset.target, test_size=0.2, random_state=42)
```

```python
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
print("R^2: ", model.score(X_test, y_test))
y_pred = model.predict(X_test)
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred))
```

```python
class MyLinearRegression:
    def __init__(self):
        self.w = None
        self.b = None
    
    def fit(self, X, y):
        n_samples, n_features = X.shape
        X_b = np.hstack((np.ones((n_samples, 1)), X))
        theta = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
        self.b = theta[0]
        self.w = theta[1:]

    def predict(self, X):
        return X.dot(self.w) + self.b

my_model = MyLinearRegression()
my_model.fit(X_train, y_train)
y_pred_my_model = my_model.predict(X_test)
print("R^2: ", r2_score(y_test, y_pred_my_model))
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred_my_model))
```

```python
from sklearn.preprocessing import StandardScaler

class LinearRegressionWithGradientDescent:
    def __init__(self, learning_rate=0.01, n_iterations=1000):
        self.learning_rate = learning_rate
        self.n_iterations = n_iterations
        self.w = None
        self.b = None

    def fit(self, X, y):
        X = np.asarray(X, dtype=float)
        y = np.asarray(y, dtype=float)

        n_samples, n_features = X.shape
        self.w = np.zeros(n_features, dtype=float)
        self.b = 0.0

        for _ in range(self.n_iterations):
            y_pred = X.dot(self.w) + self.b
            error = y_pred - y

            dw = (1 / n_samples) * X.T.dot(error)
            db = (1 / n_samples) * np.sum(error)

            self.w -= self.learning_rate * dw
            self.b -= self.learning_rate * db

    def predict(self, X):
        X = np.asarray(X, dtype=float)
        return X.dot(self.w) + self.b

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

my_model_gd = LinearRegressionWithGradientDescent(learning_rate=0.01, n_iterations=2000)
my_model_gd.fit(X_train_scaled, y_train)
y_pred_my_model_gd = my_model_gd.predict(X_test_scaled)

print("R^2: ", r2_score(y_test, y_pred_my_model_gd))
print("Mean squared error: %.2f" % mean_squared_error(y_test, y_pred_my_model_gd))
```
