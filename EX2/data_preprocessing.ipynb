```python
total_sqft_float = []
for str_val in df3['total_sqft']:
    if '-' in str_val:
        tokens = str_val.split('-')
        if len(tokens) == 2:
            try:
                total_sqft_float.append((float(tokens[0]) + float(tokens[1])) / 2)
            except ValueError:
                total_sqft_float.append(np.nan)
        else:
            total_sqft_float.append(np.nan)
    else:
        try:
            total_sqft_float.append(float(str_val))
        except ValueError:
            total_sqft_float.append(np.nan)
```

```python
df4 = df3.reset_index(drop = True)
```

```python
df5 = pd.concat([df4, pd.Series(total_sqft_float, name = 'total_sqft_float')], axis = 1)
df5.head()
```

```python
size_int = []
for str_val in df6['size']:
    try:
        size_int.append(int(str_val.split(' ')[0]))
    except ValueError:
        size_int.append(np.nan)

df6 = df6.reset_index(drop = True)
```

```python
df8 = df7[~(df7['total_sqft_float'] / df7['bhk'] < 350)]
df8.shape
```

```python
df8['price_per_sqft'] = df8['price'] * 100000 / df8['total_sqft_float']
df8.head()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('location'):
        m = np.mean(subdf.price_per_sqft)
        st = np.std(subdf.price_per_sqft)
        reduced_df = subdf[(subdf.price_per_sqft > (m-st)) & (subdf.price_per_sqft <= (m+st))]
        df_out = pd.concat([df_out, reduced_df], ignore_index=True)
    return df_out

df9 = remove_pps_outliers(df8)
df9.shape
```

```python
def remove_bhk_outliers(df):
    exclude_indices = np.array([])
    for location, location_df in df.groupby('location'):
        bhk_stats = {}
        for bhk, bhk_df in location_df.groupby('bhk'):
            bhk_stats[bhk] = {
                'mean': np.mean(bhk_df.price_per_sqft),
                'std': np.std(bhk_df.price_per_sqft),
                'count': bhk_df.shape[0]
            }

        for bhk, bhk_df in location_df.groupby('bhk'):
            stats = bhk_stats.get(bhk - 1)
            if stats and stats['count'] > 5:
                exclude_indices = np.append(exclude_indices, bhk_df[bhk_df.price_per_sqft < (stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')

df10 = remove_bhk_outliers(df9)
df10.shape
```

```python
def plot_scatter_chart(df, location):
    loc_df = df[df['location'] == location]
    plt.scatter(loc_df.total_sqft, loc_df.price, color = 'blue', marker = '+')
    plt.xlabel("Total Square Feet")
    plt.ylabel("Price")
    plt.title(location)

plot_scatter_chart(df9, "Rajaji Nagar")
```

```python
plot_scatter_chart(df9, "Hebbal")
```

```python
def statistics_price_by_area_type(df):
    area_stats = df.groupby('area_type')['price_per_sqft'].agg(['mean', 'std', 'count'])
    return area_stats
area_stats = statistics_price_by_area_type(df9)
print(area_stats)
```

```python
location_stats = (
    df9.groupby('location')['price_per_sqft']
       .agg(mean_price_per_sqft='mean', median_price_per_sqft='median', std_price_per_sqft='std', listing_count='count')
       .sort_values(by='mean_price_per_sqft', ascending=False)
)

print('Top 10 khu vuc co gia trung binh cao nhat:')
print(location_stats.head(10))

print('\nTop 10 khu vuc co gia trung binh thap nhat:')
print(location_stats.tail(10).sort_values(by='mean_price_per_sqft', ascending=True))

location_stats_10 = location_stats[location_stats['listing_count'] >= 10]

plt.figure(figsize=(12, 8))
(
    location_stats_10.head(20)
    .sort_values(by='mean_price_per_sqft', ascending=True)['mean_price_per_sqft']
    .plot(kind='barh', color='teal')
)
plt.title('Top 20 location co gia nha trung binh cao nhat (>=10 tin)')
plt.xlabel('Gia trung binh (price_per_sqft)')
plt.ylabel('Location')
plt.tight_layout()

plt.figure(figsize=(10, 5))
plt.hist(location_stats_10['mean_price_per_sqft'], bins=30, color='coral', edgecolor='black')
plt.title('Phan bo gia nha trung binh giua cac location (>=10 tin)')
plt.xlabel('Mean price_per_sqft')
plt.ylabel('So luong location')
plt.tight_layout()
```
