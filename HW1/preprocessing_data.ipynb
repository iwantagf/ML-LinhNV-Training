```python
import os
import matplotlib.pyplot as plt
import numpy as np

from sklearn.datasets import load_files
from pyvi import ViTokenizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
```

```python
INP = 'data/news_vnexpress'
os.makedirs("images", exist_ok=True)
```

```python
n = 0
for label in os.listdir(INP):
    print(f'{label}: {len(os.listdir(os.path.join(INP, label)))}')
    n += len(os.listdir(os.path.join(INP, label)))

print(f'So van ban: {n}')
data_train = load_files(container_path = INP, encoding = 'utf-8')

print("Mapping:")
for i in range(len(data_train.target_names)):
    print(f'{data_train.target_names[i]} = {i}')

print('--------------------------------------')
print(data_train.filenames[0 : 1])
print(data_train.target[0 : 1])
print(data_train.data[0 : 1])
```

```python
with open('data/vietnamese-stopwords.txt', 'r', encoding='utf-8') as f:
    stop_words = f.read().splitlines()
stop_words = [x.strip().replace(" ", "_") for x in stop_words]
print(f'So stop words: {len(stop_words)}')
print(stop_words[0 : 10])

def process_data(data_train):
    vectorizer = CountVectorizer(stop_words = stop_words)
    rf_process = Pipeline([
        ('vectorizer', vectorizer),
        ('tfidf', TfidfTransformer())
    ])

    return rf_process.fit_transform(data_train.data, data_train.target)

X = process_data(data_train)
y = data_train.target

print(f'Kich thuoc du lieu sau khi processing: {X.shape} van ban')
print(f'Kich thuoc nhan tuong ung: {y.shape}')
```
